```python
import json
import pandas as pd 
import numpy as np
from urllib.request import urlopen
```

```python
API = 'http://aflow.org/API/aflux/'
Dont_showmethis=",$catalog(ICSD),$'auid',$'spacegroup_relax',$Pearson_symbol_relax',$'aurl',$'species'"
Properties=",species(),composition(),enthalpy_formation_atom(*0),volume_cell(),density()"
MATCHBOOK=Elements+Dont_showmethis+Properties
DIRECTIVE = '$paging(0),format(json)'
REQUEST = API + '?' + MATCHBOOK + ',' + DIRECTIVE
```

```python
groups=['Metals','NonMetals']
nspecies=[2,3]
list_df=[]
for i in nspecies:
    if i==2 :
            for j in range(len(groups)):
                Elements= f'species(H,{groups[j]}),nspecies({i})'
                MATCHBOOK=Elements+Dont_showmethis+Properties
                DIRECTIVE = '$paging(0),format(json)'
                REQUEST = API + '?' + MATCHBOOK + ',' + DIRECTIVE
                response = json.loads(urlopen(REQUEST).read().decode('utf-8'))
                df=pd.DataFrame(response)
                list_df.append(df)
    elif i==3:
            for j in range(len(groups)+1):
                j1=j
                j2=j1
                if j==len(groups):
                    j1=j-1
                    j2=j1-1
                    Elements= f'species(H,{groups[j1]},{groups[j2]}),nspecies({i})'
                    MATCHBOOK=Elements+Dont_showmethis+Properties
                    DIRECTIVE = '$paging(0),format(json)'
                    REQUEST = API + '?' + MATCHBOOK + ',' + DIRECTIVE
                    response = json.loads(urlopen(REQUEST).read().decode('utf-8'))
                    df=pd.DataFrame(response)
                    list_df.append(df)
```

```python
df=pd.concat(list_df,ignore_index=True,join="inner",axis=0)
```
