```python
import  requests as rqs
import  pandas  as pd 
import numpy as np
import json
from urllib.request import urlopen
```

```python
s0='(F-Fe-Ge-Ac-Ag-Al-Am-Ar-As-At-Au-B-Ba-Be-Bh-Bi),H'
s1='(Bk-Br-C-Ca-Cd-Ce-Cf-Cl-Cm-Cn-Co-Cr-Cs-Cu-Ga-Er-Db-Ds-Dy),H'
s2='(Mg-Mn-Mo-Mt-N-Na-Nb-He-Hf-Hg-Ho-Hs-I-In-Ir-K-Kr-La-Li-Lr-Lu),H'
s3='(O-Os-P-Pb-Pd-Pm-Pr-Pt-Pu-Re)'
s4='(Xe-Y-Yb-Zn-Zr),H'
elements=[s0,s1,s2,s3,s4]
```

```python
features='formationenergy?fields=name,spacegroup,volume,element,ntypes,band_gap,delta_e&limit=2000&filter=element_set'
conditional_features='delta_e<0,ntypes<=3'
```

```python
list_df=[]
for i in range(len(elements)):
    url="http://oqmd.org/oqmdapi/"+f'{features}={elements[i]},{conditional_features}'
    response=json.loads(urlopen(url).read().decode('utf-8'))
    df=pd.DataFrame(response['data'])
    list_df.append(df)
```

```python
df=pd.concat(list_df)
```

```python
# This is the expected output. For illustrative purposes, we have reduced the value of the 'limit' parameter to 3. 
df
```

```python
df.to_csv("HydroMetalMix_oqmd_data.csv",index=True)
```
